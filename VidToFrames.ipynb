```
# Importing all necessary libraries 
import cv2 
import os 
import numpy as np
```

```
labels = os.listdir(path) # array of labels (translations of sign)
```

```
# ensure unique ids
id_list = []
for label in labels:
  filenames = os.listdir(path+label)
  for filename in filenames:
    id = filename.split('.')[0]
    if id != '':
      id_list.append(id)

# check if filenames are unique
if len(id_list) > len(set(id_list)):
  print(len(id_list))
  print(len(set(id_list)))
  print('filenames not unique')
else:
  print('filenames are unique')
```

```
def printFPS(video):
  # Find OpenCV version
  (major_ver, minor_ver, subminor_ver) = (cv2.__version__).split('.')

  if int(major_ver)  < 3 :
      fps = video.get(cv2.cv.CV_CAP_PROP_FPS)
      print("Frames per second: {0}".format(fps))
  else :
      fps = video.get(cv2.CAP_PROP_FPS)
      print("Frames per second: {0}".format(fps))
  return fps

def vidToImg(vidName, label, fps_list, dim_list):
  # Read the video from specified path 
  vid_path = path+label+'/'+vidName
  print(vid_path)
  cam = cv2.VideoCapture(vid_path) 
 
  try: 
      # creating a folder named data 
      if not os.path.exists(path+label+'/'+ vidName[0:-4]): 
        os.makedirs(path+label+'/'+ vidName[0:-4]) 
      else:
        return fps_list, dim_list
    

  # if not created then raise error 
  except OSError: 
      print('Error: Creating directory of '+vidName[0:-4]) 
    
  fps = printFPS(cam)
  fps_list[vidName[0:-4]] = fps


  # frame 
  currentframe = 0
    
  while(True): 
        
    # reading from frame 
    ret,frame = cam.read() 
  
    if ret: 
      # if video is still left continue creating images 
      name = path+label+'/'+vidName[0:-4]+ '/'+vidName[0:-4]+ '-' + str(currentframe) + '.jpg'
      frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
      
      # writing the extracted images 
      cv2.imwrite(name, frame) 

      # increasing counter so that it will 
      # show how many frames are created 
      currentframe += 1
      dims = frame.shape
    else: 
        break

  print("Dimensions:", dims)
  dim_list[vidName[0:-4]] = dims
  
  # Release all space and windows once done 
  cam.release() 
  cv2.destroyAllWindows() 
  return fps_list, dim_list
```

```
ground_truth = {}
fps_list={}
dim_list={}

labels = sorted(labels)

for label in labels:
  filenames = os.listdir(path+label)
  for filename in filenames:
    if '.' in filename:
      if filename.split('.')[1] == 'mp4':
        ground_truth[filename[0:-4]] = label
        print(filename)
        fps_list, dim_list = vidToImg(filename, label, fps_list, dim_list)
```
